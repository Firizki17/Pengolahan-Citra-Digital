# Soal operasi geometri (20 points)
Instruksi : membuat code python

Lakukan operasi geometri pada gambar, sehingga menghasilkan output gambar berikut

Hint : translasi (pergeseran) gambar sejauh 50 piksel ke kanan dan 50 piksel ke bawah.
image.png

```python
import urllib.request

urllib = urllib.request.urlretrieve(
'https://img.freepik.com/premium-vector/set-multi-colored-arrows-with-white-outline-like-stickers_655665-139.jpg', # link from google image
"gambar_4.png")

image_4 = cv2.imread('gambar_4.png')
image_rgb = cv2.cvtColor(image_4, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show()
```

```python
# Create reflections: Horizontal (flip left-right) and Vertical (flip upside down)
horizontal_reflection = cv2.flip(image_rgb, 1)  # Flip around the y-axis
# Create subplots for displaying original and reflected images
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
# Plot the original image
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')
# Plot the horizontally reflected image
axs[1].imshow(horizontal_reflection)
axs[1].set_title('Horizontal Reflection')
# Display the subplots
plt.tight_layout()
plt.show()
```

```python
# Define the translation parameters (x and y shifts)
tx = 50  # Shift 100 pixels to the right
ty = 50   # Shift 50 pixels downwards

# Create the translation matrix
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])

# Apply the translation to the image using warpAffine
translated_image = cv2.warpAffine(horizontal_reflection, translation_matrix, (image_rgb.shape[1], image_rgb.shape[0]))

# Create subplots for displaying original and translated images
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
# Plot the original image
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')
# Plot the translated image
axs[1].imshow(translated_image)
axs[1].set_title('Translation + Reflection')
# Display the subplots
plt.tight_layout()
plt.show()
```

1. Operasi Geometri

Lakukan rotasi pada gambar_2.png sebesar 30 derajat searah jarum jam dengan pusat rotasi berada di tengah gambar.

```python
import urllib.request

urllib = urllib.request.urlretrieve(
'https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcREBzFOtPbLM_FoqCKP79OGP519BQ5HuNKWDehaxUDNr3RtCBVAhCY8LXxXnFN1sM6ZZCI&usqp=CAU', # link from google image
"gambar_2.png")

image2 = cv2.imread('gambar_2.png')
image_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Assuming 'image_rgb' is already loaded (e.g., using cv2.imread)
# Get the dimensions of the image
height, width = image_rgb.shape[:2]

# Calculate the center of the image
center = (width // 2, height // 2)

# Define the rotation angle in degrees (clockwise, which is positive)
angle = 30

# Get the rotation matrix for the specified angle
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)

# Perform the rotation (keeping the same size)
rotated_image = cv2.warpAffine(image_rgb, rotation_matrix, (width, height))

# Display the rotated image using matplotlib
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for display
plt.title('Rotated Image by 30 degrees clockwise')
plt.axis('off')  # Hide axis
plt.show()
```

Lakukan translasi (pergeseran) pada gambar_3.png dengan menggeser gambar sejauh 10 piksel ke kanan dan 50 piksel ke bawah

```python
# Import the necessary libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
import requests

# URL of the image
url = "https://t4.ftcdn.net/jpg/07/81/10/11/360_F_781101127_3Wdne7Ig4t8ltgClAowBzPrfWd94bQiT.jpg"

# Download the image using requests and convert it to a numpy array
response = requests.get(url, stream=True).content
image_np = np.frombuffer(response, np.uint8)

# Decode the image using OpenCV
img = cv2.imdecode(image_np, cv2.IMREAD_COLOR)

# Convert BGR image to RGB for plotting
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Get the image dimensions (height, width)
width = image_rgb.shape[1]
height = image_rgb.shape[0]

# Define translation parameters
tx = 10  # Shift in x-direction (to the right)
ty = 70   # Shift in y-direction (down)

# Translation matrix
translation_matrix = np.array([[1, 0, tx], [0, 1, ty]], dtype=np.float32)

# Apply the translation using warpAffine
translated_image = cv2.warpAffine(image_rgb, translation_matrix, (width, height))

# Create subplots for displaying original and translated images
fig, axs = plt.subplots(1, 2, figsize=(7, 4))

# Plot the original image
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')

# Plot the translated image
axs[1].imshow(translated_image)
axs[1].set_title('Image Translation')

# Remove ticks from the subplots
for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

# Display the subplots
plt.tight_layout()
plt.show()
```

2. Operasi Aritmatika

Gabungkan kedua gambar hasil operasi geometri menggunakan penjumlahan (cv2.add).

```python
# plot figure
plt.figure(figsize=(8,4))

plt.subplot(1, 2, 1)
plt.imshow(rotated_image)
plt.title("rotated_image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(translated_image)
plt.title("translated_image")
plt.axis('off')

plt.show()
```

```python
# hasil gambar Addition
# Resize rotated_image to match the dimensions of translated_image
rotated_image = cv2.resize(rotated_image, (translated_image.shape[1], translated_image.shape[0]))

result_add = cv2.add(rotated_image, translated_image)
plt.imshow(result_add)
plt.show()

# Dapatkan resolusi gambar dalam piksel
width, height,channel = result_add.shape
print(f"Resolusi gambar 1: {width}x{height}x{channel} piksel")
```

3. Operasi Logika

Lakukan operasi NOT (cv2.bitwise_not) pada hasil penjumlahan untuk menghasilkan gambar akhir.

```python
# hasil gambar Addition
Gambar_akhir = cv2.bitwise_not(rotated_image, translated_image,mask = None)
plt.imshow(Gambar_akhir)
plt.show()


# Dapatkan resolusi gambar dalam piksel
width, height,channel = Gambar_akhir.shape
print(f"Resolusi gambar 1: {width}x{height}x{channel} piksel")
```

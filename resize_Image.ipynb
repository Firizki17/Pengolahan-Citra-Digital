```python
# prompt: load image using urllib

import urllib.request
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

image_url = "https://images.unsplash.com/photo-1609668233461-019a1e3188ec?fm=jpg&q=60&w=3000&ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D.gfg.png"  # Replace with your image URL

try:
  with urllib.request.urlopen(image_url) as url:
    img = mpimg.imread(url, format='gfg.png')
    plt.imshow(img)
    plt.show()
except Exception as e:
  print(f"Error loading image: {e}")
```

```python
# prompt: resize image using opencv resulution 736*736 and saving

!pip install opencv-python

import urllib.request
import cv2
import numpy as np

image_url = "https://images.unsplash.com/photo-1609668233461-019a1e3188ec?fm=jpg&q=60&w=3000&ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D.gfg.png"

try:
    with urllib.request.urlopen(image_url) as url:
        # Read image data from URL
        image_array = np.asarray(bytearray(url.read()), dtype=np.uint8)
        # Decode image using OpenCV
        img = cv2.imdecode(image_array, cv2.IMREAD_COLOR)

        # Resize image to 736x736
        resized_img = cv2.resize(img, (736, 736))

        # Save the resized image
        cv2.imwrite("resized_image.jpg", resized_img)
        print("Resized image saved as resized_image.jpg")

except Exception as e:
    print(f"Error loading or processing image: {e}")

```

```python
# library
import urllib.request
import cv2 # make sure cv2 is imported
import numpy as np
import matplotlib.pyplot as plt

# first Image
urllib.request.urlretrieve(
'https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQXEYMrwPKZZpYLtuznkmaOTEYILF5yw-mwaA&s.png', # link from google image
"gambar_1.png")

gambar_1 = cv2.imread('gambar_1.png', 0)

# load the image using cv2
gambar_1 = cv2.imread('gambar_1.png')

gambar_1 = cv2.cvtColor(gambar_1, cv2.COLOR_BGR2RGB) # convert gambar_1 to RGB
```

```python
plt.imshow(gambar_1)
```

```python
# find frequency of pixels in range 0-255
histr = cv2.calcHist([gambar_1],[0],None,[256],[0,256])
# show the plotting graph of an image
plt.plot(histr)
plt.show()
```

```python
# calculate frequency of pixels in range 0-255
histg = cv2.calcHist([gambar_1],[0],None,[256],[0,256])
```

```python
# alternative way to find histogram of an image
plt.hist(gambar_1.ravel(),256,[0,256])
plt.show()
```

```python
# creating a Histograms Equalization
# of a image using cv2.equalizeHist()
# convert gambar_1 to grayscale
gray = cv2.cvtColor(gambar_1, cv2.COLOR_BGR2GRAY)

# apply histogram equalization to the grayscale image
equ = cv2.equalizeHist(gray)

# stacking images side-by-side
res = np.hstack((gray, equ)) # Updated to use gray image

# show image input vs output
plt.imshow(res, cmap='gray') # Updated to use gray colormap

plt.show()
```
